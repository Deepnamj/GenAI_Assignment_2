Task 1: Discount Rules (if / elif / else)

1. Write a program that reads an integer order_amount from the user using input().

2. Apply the following discount rules and print the final amount:

    order_amount >= 2000 → 15% discount

    1500 <= order_amount < 2000 → 10% discount
    
    1000 <= order_amount < 1500 → 7% discount
    
    otherwise → 0% discount

3. Ensure you convert input to a number and handle non-numeric input with an error message and exit.

4. Add tax (fixed 5%) after discount and print subtotal, tax, and final total.

```python
# Get order amount from user
user_input = input("Enter the order amount: ")

# Check if the input contains only digits
if not user_input.isdigit():
    print("Invalid input. Please enter a numeric value.")
    exit()

# Convert input to integer
order_amount = int(user_input)

# Apply discount based on order amount
if order_amount >= 2000:
    discounted_amount = order_amount * 0.15
elif 1500 <= order_amount < 2000:
    discounted_amount = order_amount * 0.10
elif 1000 <= order_amount < 1500:
    discounted_amount = order_amount * 0.07
else:
    discounted_amount = 0

# Calculate amount after discount
final_amount = order_amount - discounted_amount

print("Original Amount:", order_amount)
print("Discount Applied:", int(discounted_amount))
print("Subtotal (after discount):", int(final_amount))

# Calculate 5% tax
tax = final_amount * 0.05

print("Tax (5%):", tax)
print("Final Total:", final_amount + tax)
```

Task 2: Process Multiple Orders (for loop)

1. Given orders = [1200, 2500, 800, 1750, 3000], use a for loop to apply the discount rules from Task 1 to each order and print a summary table: 
    
    order_amount -> discount% -> final_amount.

2. Compute and print the total revenue after discounts.

3. Print the number of orders that received a discount (discount > 0).

```python
# List of order amounts
orders = [1200, 2500, 800, 1750, 3000]

# Variables to track total revenue and number of discounted orders
total_revenue = 0
discounted_orders = 0

# Loop through each order
for order in orders:

        # Determine discount based on order value
        if order >= 2000:
            discount = order * 0.15
        elif 1500 <= order < 2000:
            discount = order * 0.10
        elif 1000 <= order < 1500:
            discount = order * 0.07
        else:
            discount = 0

        # Count orders that received a discount
        if discount > 0:
            discounted_orders += 1

        # Calculate final amount after discount
        final_amount = order - discount

        # Calculate total revenue
        total_revenue += final_amount

        # Print order, discount percentage, and final amount
        print(f"{order} -> {discount/order*100:.0f}% -> {final_amount:.2f}")

# Print total revenue after all discounts
print("Total revenue after discounts:", total_revenue)

# Print number of orders that received discounts
print("Number of orders with discount:", discounted_orders)
```

Task 3: User Menu (while loop + break/continue)

1. Implement a while loop driven menu that keeps asking the user to choose an action until they press q to quit. Menu options:

    1 — Add order amount to a running list (use input())

    2 — Show all orders and totals after applying discounts
    
    q — Quit

2. Use continue to re-show the menu after invalid input and break to exit on q.
Constraints: Keep the loop logic simple and use only lists and loops (no functions required).

```python
# List to store all order amounts
orders = []

# Infinite loop to keep the menu running
while True:
    print("Menu:")
    print("1. Add new order amount")
    print("2. All orders and totals after discount") 
    print("3. Press q - To Quit")

    # Get user choice
    choice = input("Enter your choice: ")

    # Option 1: Add a new order
    if choice == '1':
        new_order_amount = input("Enter the new order amount: ")

        # Validate numeric input
        if new_order_amount.isdigit() == False:
            print("Invalid input. Please enter a numeric value.")
            continue

        # Convert to integer and add to orders list
        new_order = int(new_order_amount)
        orders.append(new_order)
        print(f"New order of {new_order} added.")

    # Option 2: Display orders with discounts and totals
    elif choice == '2':
        total_revenue = 0
        discounted_orders = 0

        # Process each order
        for order in orders:

            # Apply discount based on order amount
            if order >= 2000:
                discount = order * 0.15
            elif 1500 <= order < 2000:
                discount = order * 0.10
            elif 1000 <= order < 1500:
                discount = order * 0.07
            else:
                discount = 0

            # Calculate final amount after discount
            final_amount = order - discount
            total_revenue += final_amount

            # Display order details
            print(f"{order} -> {discount/order*100:.0f}% -> {final_amount:.2f}")

        # Print summary
        print("Total revenue after discounts:", f"${total_revenue:.2f}")

    # Option to quit the program
    elif choice == 'q':
        print("Exiting the program.")   
        break
    
    # Handle invalid menu choices
    else:
        print("Invalid choice. Please try again.")
```

Task 4: Loop Control with Conditions (break & continue)

1.Given daily = [200, 150, 0, 400, 50, -1, 300], iterate through it with a for loop and:

    If a day's value is -1, treat it as corrupted data and break (stop processing).
    If a day's value is 0, treat it as a day with no sales and continue (skip adding to revenue).
    For valid positive sales, add to total_sales and print the running total.

2. Print the final total after the loop completes (or stops due to corrupted data).

```python
# List containing daily sales data
daily = [200, 150, 0, 400, 50, -1, 300]

# Variable to store total sales
total_sales = 0 

# Loop through each day's sales
for i in range(len(daily)):

    # Stop processing if corrupted data (-1) is found
    if daily[i] == -1:
        print("Corrupted data detected. Stopping.")
        break

    # Skip the day if there were no sales
    if daily[i] == 0:
        print(f"Day {i+1}: No sales today.")
        continue
    else:
        # Add sales to the running total
        total_sales += daily[i]   
        print(f"Day {i+1}: Sales = {daily[i]} | Running Total = {total_sales}")

# Print final total sales
print(f"\nFinal Total Sales: {total_sales}")
```
