Task 1: Discount Rules (if / elif / else)

1. Write a program that reads an integer order_amount from the user using input().

2. Apply the following discount rules and print the final amount:

    order_amount >= 2000 → 15% discount

    1500 <= order_amount < 2000 → 10% discount
    
    1000 <= order_amount < 1500 → 7% discount
    
    otherwise → 0% discount

3. Ensure you convert input to a number and handle non-numeric input with an error message and exit.

4. Add tax (fixed 5%) after discount and print subtotal, tax, and final total.

```python
user_input = input("Enter the order amount: ")

if not user_input.isdigit():
    print("Invalid input. Please enter a numeric value.")
    exit()

order_amount = int(user_input)

if order_amount >= 2000:
    discounted_amount = order_amount * 0.15
elif 1500 <= order_amount < 2000:
    discounted_amount = order_amount * 0.10
elif 1000 <= order_amount < 1500:
    discounted_amount = order_amount * 0.07
else:
    discounted_amount = 0

final_amount = order_amount - discounted_amount

print("Original Amount:", order_amount)
print("Discount Applied:", int(discounted_amount))
print("Subtotal (after discount):", int(final_amount))

tax = final_amount * 0.05
print("Tax (5%):", tax)
print("Final Total:", final_amount + tax)
```
